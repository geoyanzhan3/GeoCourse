### What is Matplotlib

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python (https://matplotlib.org).



#### Create a line/scatter plot

numpy and matplotlib modules are used to create plots

```python
import numpy as np
import matplotlib.pyplot as plt

# create the array x
x = np.linspace(0,10,101)
# create the function y
y = x**2 + 2*x + 1

# plot with red points
plt.plot(x, y, 'r.')
plt.show()
```

```python
# plot with blue line
plt.plot(x, y, 'b-')
# adding labels
plt.xlabel('x')
plt.ylabel('y')
# adding title
plt.title('y = x^2+2x+1')
plt.show()
```

```python
# multiple subplots
# specify figure number and size
fig = plt.figure(1, figsize=(4,6))
# subplots
plt.subplot(2,1,1)
plt.plot(x, y, 'b-')
# subplots
plt.subplot(2,1,2)
plt.plot(x, y, 'r--')
plt.show()
fig.savefig("plot1.png")
```

#### Histogram

```python
import numpy as np
import matplotlib.pyplot as plt

# create a random uniform distribution
uniform_dis = np.random.rand(1000)
# create a random normal distribution
normal_dis = np.random.randn(1000)

## plot
# specify figure number and size
fig = plt.figure(2, figsize=(4,7))
# subplots
plt.subplot(2,1,1)
plt.hist(uniform_dis, 20)
plt.title('Uniform Distribution')
# subplots
plt.subplot(2,1,2)
plt.hist(normal_dis, 20)
plt.title('Normal Distribution')
plt.show()
fig.savefig("plot2.png")
```

#### 2D image

```python
import numpy as np
import matplotlib.pyplot as plt

# create an x,y grid
xi = np.linspace(-10,10,101)
yi = np.linspace(-10,10,101)
# create a meshgrid
X, Y = np.meshgrid(xi, yi)

# create a 2D surface
Z = (X**2 + Y**2) * np.sin(X + Y)

# plot
plt.imshow(Z, extent=[-10, 10, -10, 10], origin='lower')
plt.colorbar()
plt.show()
```

```python
# change colormap and color-range
# import module for the colormap
import matplotlib.cm as cm

# plot
plt.imshow(Z, extent=[-10, 10, -10, 10], 
           origin='lower',
           cmap=cm.RdYlGn,
           vmin=-250,
           vmax=250
          )
plt.colorbar(shrink=0.5)
plt.show()
```

#### Exercise: plot the Mogi (1958) model

hint: the surface displacements can be expressed by

$$u = \frac{\alpha x}{(d^2 + R^2)^ {\frac{3}{2}}} \tag{1}$$

$$v = \frac{\alpha y}{(d^2 + R^2)^ {\frac{3}{2}}} \tag{2}$$

$$w = \frac{\alpha d}{(d^2 + R^2)^ {\frac{3}{2}}} \tag{3}$$

In which, $\alpha= a^3 p \frac{(1 - \nu)}{\mu}$, and $R = \sqrt{x^2 + y^2}$.

where $d$ is the depth of the chamber center, $a$ is the radius of chamber, and $p$ is the overpressure along the chamber wall. For the material properties, $\mu$ is shear modulus and $\nu$ is Poisson's Ratio.


#### 3D plot


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

# create an x,y grid
xi = np.linspace(-10,10,51)
yi = np.linspace(-10,10,51)
# create a meshgrid
X, Y = np.meshgrid(xi, yi)

# create a 2D surface
Z = (X**2 + Y**2) * np.sin(X + Y)

# plot
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.viridis)
ax.view_init(elev=60, azim=15)
plt.show()
```
